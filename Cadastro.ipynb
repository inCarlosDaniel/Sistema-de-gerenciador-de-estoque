```python
%%writefile Cadastro.py

import datetime
import math
import random


def ler_numero_positivo(mensagem):  # Funcao para ler um numero positivo do usuario, com validacao
    
    valor = float(input(mensagem).replace(",", ".")) # Substitui virgula por ponto para permitir entrada de numeros decimais com virgula

    if not math.isfinite(valor) or valor < 0: # Verifica se o valor e finito e positivo
        
        raise ValueError("Digite um numero valido e positivo.") # Lanca um erro se o valor for invalido

    return valor # Retorna o valor lido


def gerar_identificacao(): # Funcao para gerar uma identificacao unica para o produto
    
    return f"PROD-{random.randint(100000, 999999)}" # Retorna uma string com o formato "PROD-XXXXXX" onde XXXXXX e um numero aleatorio de 6 digitos


def cadastrar_produto(estoque, produto): # Funcao para cadastrar ou atualizar um produto no estoque
    
    nome = produto["nome"] # Extrai o nome do produto do dicionario recebido como parametro

    if nome in estoque: # Verifica se o produto ja existe no estoque
        
        estoque[nome]["quantidade"] += produto["quantidade"] 
        
        estoque[nome].update({
            
            "preco": produto["preco"],
            
            "peso": produto["peso"],
            
            "identificacao": produto["identificacao"],
            
            "espe_tecnica": produto["espe_tecnica"],
            
            "data_inicio": produto["data_inicio"],
            
            "data_cadastro": produto["data_cadastro"],
            
        })
        
        print(f"Produto '{nome}' atualizado com sucesso.")
        
    else:
        
        estoque[nome] = produto
        
        print(f"Produto '{nome}' adicionado ao estoque com sucesso.")
        


def solicitar_dados_produto():
    
    print("\nCadastro de Produtos")
    
    nome = input("Digite o nome do produto: ").strip()
    
    preco = ler_numero_positivo("Digite o preco do produto: ")
    
    peso = ler_numero_positivo("Digite o peso do produto: ")
    
    quantidade = int(input("Digite a quantidade em estoque: "))
    
    identificacao = input("Digite a identificacao do produto (enter para gerar): ").strip()

    if not identificacao:  
        
        identificacao = gerar_identificacao() 
    
    espe_tecnica = input("Digite a especificacao tecnica do produto: ").strip()
    
    data_inicio = input("Digite a data de inicio do produto (enter para hoje): ").strip()

    if not data_inicio:
        
        data_inicio = datetime.date.today().strftime("%d/%m/%Y") # Formata a data atual no formato"

    data_cadastro = datetime.datetime.now().strftime("%d/%m/%Y %H:%M:%S") # Formata a data e hora atual no formato "

    return {
        
        "nome": nome,
        
        "preco": preco,
        
        "peso": peso,
        
        "quantidade": quantidade,
        
        "identificacao": identificacao,
        
        "espe_tecnica": espe_tecnica,
        
        "data_inicio": data_inicio,
        
        "data_cadastro": data_cadastro,
        
    }


def exibir_estoque(estoque):
    
    print("\nEstoque atual:")
    
    if not estoque:
        
        print("Nenhum produto cadastrado.")
        
        return

    for nome, dados in estoque.items():
        
        print(f"Nome: {nome}")
        
        print(f"Preco: R${dados['preco']:.2f}")
        
        print(f"Peso: {dados['peso']}")
        
        print(f"Quantidade: {dados['quantidade']}")
        
        print(f"Identificacao: {dados['identificacao']}")
        
        print(f"Especificacao tecnica: {dados['espe_tecnica']}")
        
        print(f"Data de inicio: {dados['data_inicio']}")
        
        print(f"Data de cadastro: {dados['data_cadastro']}")
        
        print("-" * 30)

estoque = {}

while True:
    
    produto = solicitar_dados_produto()
    
    cadastrar_produto(estoque, produto)
    
    exibir_estoque(estoque)

    continuar = input("Deseja cadastrar outro produto? [s/N]: ").strip().lower()
    
    if continuar not in ("s", "sim"):
        
        break

```
