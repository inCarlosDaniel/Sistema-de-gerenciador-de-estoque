```python
import sqlite3 # Importa a biblioteca para trabalhar com SQLite

conexao = sqlite3.connect("estoque.db") # Conecta ao banco de dados estoque.sqlite3. Se o arquivo não existir, ele será criado.

cursor = conexao.cursor() # Cria um cursor para executar comandos SQL

cursor.execute("""
CREATE TABLE IF NOT EXISTS produtos (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    nome TEXT NOT NULL,
    preco REAL NOT NULL,
    peso REAL NOT NULL,
    quantidade INTEGER NOT NULL,
    identificacao TEXT NOT NULL UNIQUE,
    espe_tecnica TEXT,
    data_inicio TEXT NOT NULL
)
""")

conexao.commit()


```

```python
cursor.execute("SELECT name FROM sqlite_master WHERE type='table'")
cursor.fetchall()
```

```python
cursor.execute("""
INSERT INTO produtos (
    nome, preco, peso, quantidade, identificacao, espe_tecnica, data_inicio
)
VALUES (?, ?, ?, ?, ?, ?, ?)
""", (
    "Maçã",
    3.50,
    0.2,
    10,
    "001",
    "Fruta vermelha",
    "2026-05-12"
))

conexao.commit()
```

```python
cursor.execute("SELECT * FROM produtos")
cursor.fetchall()
```

```python
conexao.close() # Fecha a conexão com o banco de dados
```
